```python
orders_schema = Schema(
    {
    #"_id":ObjectId,
    "Row ID":pyarrow.int64(),
    'Order ID':pyarrow.string(), 
    'Order Date':pyarrow.int64(), 
    'Ship Date':pyarrow.int64(), 
    'Ship Mode':pyarrow.string(),
    'Customer ID':pyarrow.string(), 
    'Customer Name':pyarrow.string(), 
    'Segment':pyarrow.string(), 
    'City':pyarrow.string(), 
    'State':pyarrow.string(), 
    'Country':pyarrow.string(),
    'Postal Code':pyarrow.int64(), 
    'Market':pyarrow.string(), 
    'Region':pyarrow.string(), 
    'Product ID':pyarrow.string(), 
    'Category':pyarrow.string(),
    'Sub-Category':pyarrow.string(), 
    'Product Name':pyarrow.string(), 
    'Sales':pyarrow.int64(), 
    'Quantity':pyarrow.int64(), 
    'Discount':pyarrow.int64(),
    'Profit':pyarrow.float64(), 
    'Shipping Cost':pyarrow.int64(), 
    'Order Priority':pyarrow.string()
})
```

```python
pipelane1 = [{
        "$match": {
            "Ship Mode":"Same Day"
        }
    }]
pipelane2 = [{
    "$match": {
        "$and":[ #$and -> $or
                {"Ship Mode":"Same Day"},
                {"Category":{"$ne":"Furniture"}},
                {"Profit":{"$lt":0}}, # $gt -> $gte, $lt, $lte                
                {"Order Priority": {"$in":["High", "Critical"]}} # $in -> $nin (not in)
            ]   
        }
    }]
```

```python
orders.aggregate_pandas_all(pipelane2, schema=orders_schema)
```
